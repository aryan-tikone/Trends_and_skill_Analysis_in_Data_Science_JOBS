```python
import pandas as pd

# Load datasets
postings_df = pd.read_csv('job_postings.csv')
skills_df = pd.read_csv('job_skills.csv')
summary_df = pd.read_csv('job_summary.csv')

# Check data quickly
print(postings_df.shape, skills_df.shape, summary_df.shape)

```

```python
print(postings_df.columns)
print(skills_df.columns)
print(summary_df.columns)
```

```python
# Some jobs might have multiple rows with the same job_link
skills_df = skills_df.groupby('job_link')['job_skills'].apply(lambda x: ', '.join(x.dropna())).reset_index()
```

```python
# Merge job_postings + job_skills
merged_df = postings_df.merge(skills_df, on='job_link', how='left')

# Merge with summary
merged_df = merged_df.merge(summary_df, on='job_link', how='left')
```

```python
# Drop irrelevant columns
drop_cols = ['last_processed_time', 'last_status', 'got_summary',
             'got_ner', 'is_being_worked', 'first_seen']
# merged_df.drop(columns=drop_cols, inplace=True)

# Standardize text fields
text_cols = ['job_title', 'company', 'job_location', 'job_skills', 'job_summary']
for col in text_cols:
    merged_df[col] = merged_df[col].astype(str).str.lower().str.strip()


# Reset index after filtering
merged_df.reset_index(drop=True, inplace=True)

# Save cleaned data
merged_df.to_csv('cleaned_job_data.csv', index=False)
print("Cleaned dataset saved as cleaned_job_data.csv")
print("Final shape:", merged_df.shape)
```

Day 6: Skill Extraction Using NLP

```python
import spacy
from collections import Counter

# Load English NLP model
nlp = spacy.load("en_core_web_sm")

# Sample skill list
skill_keywords = ['python', 'sql', 'excel', 'machine learning', 'deep learning', 'tableau',
                  'power bi', 'pandas', 'numpy', 'matplotlib', 'tensorflow', 'nlp', 'spark']

def extract_skills(text):
    found = []
    if isinstance(text, str):
        for skill in skill_keywords:
            if skill in text:
                found.append(skill)
    return list(set(found))

merged_df['extracted_skills'] = merged_df['job_summary'].apply(extract_skills)

# Frequency of skills
skill_counter = Counter([skill for sublist in merged_df['extracted_skills'] for skill in sublist])
top_skills = dict(skill_counter.most_common(15))

# WordCloud
from wordcloud import WordCloud
wc = WordCloud(width=800, height=400).generate_from_frequencies(top_skills)

plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.title("Most Mentioned Skills in Job Summaries")
plt.show()
```

```python
def clean_job_level(level):
    if pd.isnull(level):
        return 'unknown'
    
    level = level.lower().strip()
    
    if 'associate' in level or 'mid' in level:
        return 'mid'
    elif 'senior' in level or 'lead' in level:
        return 'senior'
    elif 'entry' in level or 'intern' in level:
        return 'entry'
    else:
        return 'other'

merged_df['job_level_clean'] = merged_df['job_level'].apply(clean_job_level)
print(merged_df['job_level_clean'].value_counts())

```

```python
print(merged_df['job_level_clean'].value_counts())
```

```python
# Explode the skills
exploded = merged_df.explode('extracted_skills')
exploded = exploded[exploded['extracted_skills'].notnull()]

# Group by skill and job level
skill_by_level = (
    exploded.groupby(['extracted_skills', 'job_level_clean'])
    .size()
    .unstack(fill_value=0)
)

# Check available columns (job levels)
print(skill_by_level.columns)  # Should include 'mid'

# Plot top skills for mid-level
if 'mid' in skill_by_level.columns:
    top_skills_mid = skill_by_level.sort_values('mid', ascending=False).head(10)

    top_skills_mid.plot(kind='bar', figsize=(12,6))
    plt.title("Top Skills in Mid-Level Jobs")
    plt.ylabel("Frequency")
    plt.xlabel("Skills")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No 'mid' level jobs found.")

```

 Day 8: Skill Demand by Company

```python
# Filter top companies
top5_companies = merged_df['company'].value_counts().head(5).index.tolist()

# Filter and explode
company_df = merged_df[merged_df['company'].isin(top5_companies)].explode('extracted_skills')

# Pivot
skill_company = company_df.groupby(['company', 'extracted_skills']).size().unstack().fillna(0)

# Plot
skill_company.T.sort_values(by=top5_companies, ascending=False).head(10).plot(kind='bar', figsize=(12,6))
plt.title("Top Skills by Top 5 Companies")
plt.ylabel("Frequency")
plt.xlabel("Skill")
plt.show()
```
